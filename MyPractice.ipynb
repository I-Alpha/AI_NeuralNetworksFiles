```
import numpy as np
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot
```

```
mymodel = keras.Sequential([
    keras.layers.Dense(units = 1, input_shape = [1]),
    keras.layers.Dense(units = 1),
   
    
    
    ])

mymodel.compile(optimizer = 'SGD', loss = 'mean_squared_error')       
```

```
xs = np.array ([-2.0,-1.0,0.0,1.0,2.0,3.0], dtype =float)
ys = np.array ([9.0,16.0,25.0,36.0,49.0,64.0], dtype =float)
history = mymodel.fit(xs,ys, epochs = 500) 


```

```
print(mymodel.predict([-1.0]))
```

```
pyplot.title('Loss')
pyplot.plot(history.history['loss'], label='train')
pyplot.legend()
pyplot.show()
```
